```python
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import time
import datetime
```

```python
# Set the API endpoint and query parameters
url = 'https://api.opendota.com/api/publicMatches'
params = {'mmr_descending': 'true', 'limit': '100'}

data_match = []

# Define a function to fetch data in batches of 100, with a delay between batches
def fetch_data():
    num_matches = 100000
    while len(data_match) < num_matches:
        offset = len(data_match)
        params['offset'] = offset
        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()
            # if no more data available, break loop
            if not data:  
                break
            data_match.extend(data)
        else:
            print(f"Error fetching data from API: {response.status_code} - {response.reason}")
            break
        
        # total number of matches fetched so far
        progress = offset + len(data)  
        # percentage of matches fetched
        percentage = progress / num_matches  
        # number of matches remaining to fetch
        remaining = num_matches - progress  
        # current time
        now = datetime.datetime.now()  
        # time elapsed since start of function
        elapsed = (now - start_time).total_seconds()  
        # rate of matches fetched per second
        rate = progress / elapsed if elapsed > 0 else 0  
        # estimated time of finishing
        eta = datetime.datetime.now() + datetime.timedelta(seconds=(remaining / rate))  
        
        print(f"Fetched {progress} matches ({percentage:.2%}). {remaining} matches remaining. Rate: {rate:.2f} matches/second. ETA: {eta}.")
        
        time.sleep(1)  # delay between requests

# Store the start time of the function
start_time = datetime.datetime.now()

# Call the fetch_data function to fetch all the data
fetch_data()

# Convert the list data_match to a NumPy array
data_match_array = np.array(data_match)

# Print the shape of the data_match_array
print(data_match_array.shape)
```

```python
#convert to dataframe
df = pd.json_normalize(data_match_array)
df
```

```python
#drop unnecessary columns
df = df.drop(['match_seq_num', 'start_time', 'duration', 'num_mmr', 'lobby_type', 'game_mode', 'avg_rank_tier', 'num_rank_tier', 'cluster'], axis=1)
df
```

```python
#write csv
df.to_csv('dota2_data.csv')
```
